## Generate Traffic Accident Data

]Create a synthetic dataset for traffic accidents, including columns for road conditions, weather, time of day, location (latitude and longitude), and accident severity. This step will fulfill the request to 'create a dataset'.


```python
import pandas as pd
import numpy as np

print("Libraries pandas and numpy imported successfully.")
```

```python
num_accidents = 1000

road_conditions = ['Dry', 'Wet', 'Icy', 'Snowy']
weather_conditions = ['Clear', 'Rain', 'Fog', 'Snow', 'Windy']
time_of_day = ['Morning', 'Afternoon', 'Evening', 'Night']
accident_severity = ['Minor', 'Moderate', 'Severe', 'Fatal']

road_conditions_data = np.random.choice(road_conditions, num_accidents)
weather_data = np.random.choice(weather_conditions, num_accidents)
time_of_day_data = np.random.choice(time_of_day, num_accidents)
accident_severity_data = np.random.choice(accident_severity, num_accidents)

latitude_data = np.random.uniform(33.7, 34.3, num_accidents)
longitude_data = np.random.uniform(-118.6, -117.9, num_accidents)

data = {
    'Road_Conditions': road_conditions_data,
    'Weather': weather_data,
    'Time_of_Day': time_of_day_data,
    'Latitude': latitude_data,
    'Longitude': longitude_data,
    'Accident_Severity': accident_severity_data
}

df_accidents = pd.DataFrame(data)

print(f"Synthetic dataset with {num_accidents} accidents created successfully.")
print("First 5 rows of the dataset:")
print(df_accidents.head())
```

## Analyze Road Condition Patterns

Analyze the generated traffic accident data to identify patterns and correlations between various road conditions and accident frequency or severity.


```python
road_condition_frequency = df_accidents['Road_Conditions'].value_counts()
road_condition_severity_distribution = pd.crosstab(df_accidents['Road_Conditions'], df_accidents['Accident_Severity'])

print("Frequency of accidents by Road Condition:")
print(road_condition_frequency)
print("\nDistribution of Accident Severity by Road Condition:")
print(road_condition_severity_distribution)
```

## Analyze Weather Condition Patterns


Analyze the generated traffic accident data to identify patterns and correlations between various weather conditions and accident frequency or severity.

```python
weather_condition_frequency = df_accidents['Weather'].value_counts()
weather_condition_severity_distribution = pd.crosstab(df_accidents['Weather'], df_accidents['Accident_Severity'])

print("Frequency of accidents by Weather Condition:")
print(weather_condition_frequency)
print("\nDistribution of Accident Severity by Weather Condition:")
print(weather_condition_severity_distribution)
```

## Analyze Time of Day Patterns

Analyze the generated traffic accident data to identify patterns and correlations between the time of day and accident frequency or severity.

```python
time_of_day_frequency = df_accidents['Time_of_Day'].value_counts()
time_of_day_severity_distribution = pd.crosstab(df_accidents['Time_of_Day'], df_accidents['Accident_Severity'])

print("Frequency of accidents by Time of Day:")
print(time_of_day_frequency)
print("\nDistribution of Accident Severity by Time of Day:")
print(time_of_day_severity_distribution)
```

## Visualize Accident Hotspots

Generate a visualization, such as a scatter plot or heatmap, to display geographical areas with a high concentration of accidents (accident hotspots) using the latitude and longitude data. Ensure to provide appropriate legends.


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_accidents, x='Longitude', y='Latitude', hue='Accident_Severity', palette='viridis', alpha=0.7)
plt.title('Traffic Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Accident Severity')
plt.grid(True)
plt.show()

print("Accident hotspots visualization created successfully.")
```
